Q2. What is the difference between the mean, median, and mode? How are they used to measure the 
central tendency of a dataset?

The mean, median, and mode are statistical measures used to describe the central tendency of a dataset, which is a way of summarizing the "typical" or central value of a distribution of data. 

The mean, also known as the average, is calculated by summing up all the values in a dataset and dividing the sum by the total number of values. It is affected by extreme values and outliers, as it takes into account the magnitude of each data point. The mean is commonly used when the data is normally distributed or symmetrical.

The median is the middle value in a dataset when it is arranged in ascending or descending order. If there is an even number of values, the median is the average of the two middle values. The median is less affected by extreme values and outliers compared to the mean, making it a useful measure when there are outliers or the data is skewed.

The mode is the value that appears most frequently in a dataset. A dataset can have no mode (if no value appears more than once) or multiple modes (if multiple values have the same highest frequency). The mode is useful for categorical or discrete data, as it identifies the most common category or value.

Each of these measures provides different insights into the central tendency of a dataset and can be used depending on the characteristics of the data. The mean is often used in cases where the data is normally distributed and there are no significant outliers. The median is preferred when the data has outliers or is skewed, as it provides a more robust estimate of central tendency. The mode is useful for identifying the most typical category or value in categorical data. It's important to consider the nature of the data and the research question at hand when selecting an appropriate measure of central tendency.

Q10. Explain the difference between covariance and correlation. How are these measures used in 
statistical analysis

Covariance:
Covariance measures how two variables vary together. It quantifies the degree and direction of the linear relationship between two variables.

Correlation:
Correlation is a standardized measure that quantifies both the strength and direction of the linear relationship between two variables. 

In statistical analysis, covariance and correlation are both used to assess the relationship between variables. Some common applications include:

1. Descriptive Analysis: Covariance and correlation can be used to summarize and describe the relationship between variables in a dataset, providing insights into how they change together.

2. Variable Selection: Correlation is often used in feature selection to identify highly correlated variables and avoid multicollinearity in regression models.

3. Portfolio Analysis: Covariance and correlation are essential in finance for analyzing the relationships between different assets in a portfolio, assessing diversification, and managing risk.

4. Research and Hypothesis Testing: Covariance and correlation can be used to test hypotheses and evaluate the strength and significance of relationships between variables.

where x₁, x₂, x₃, ..., xn are the individual values in the dataset, and n is the total number of values in the dataset.

Q13. How is covariance different from correlation?

Covariance and correlation are both statistical measures that describe the relationship between two variables, but they have some key differences:

1. Scale Dependence: Covariance is a measure of the extent to which two variables vary together, but it is affected by the scale of the variables. This means that the magnitude of the covariance value depends on the units of measurement of the variables. Correlation, on the other hand, is a standardized measure that is scale-independent. It normalizes the covariance by dividing it by the product of the standard deviations of the variables, resulting in a value between -1 and 1.

2. Interpretability: Covariance is measured in the units of the variables being analyzed, which makes it difficult to compare the strength of the relationship across different datasets or variables with different scales. Correlation, being scale-independent and ranging between -1 and 1, allows for easier interpretation and comparison of the strength and direction of the relationship.

3. Range and Direction: Covariance can take any value, positive or negative, depending on the direction of the relationship. A positive covariance indicates a positive relationship (when one variable increases, the other tends to increase), while a negative covariance indicates a negative relationship (when one variable increases, the other tends to decrease). Correlation, however, ranges from -1 to 1, where -1 represents a perfect negative relationship, 1 represents a perfect positive relationship, and 0 represents no linear relationship.

4. Standardization: Covariance is not standardized, which means its value depends on the scale of the variables. Correlation, on the other hand, standardizes the covariance, making it comparable and allowing for a clear interpretation of the strength and direction of the relationship.


Q14. How do outliers affect measures of central tendency and dispersion? Provide an example

Outliers can have a significant impact on measures of central tendency and dispersion, affecting their values and interpretation. Here's how outliers can affect these measures:

Measures of Central Tendency (e.g., Mean, Median, Mode):
1. Mean: Outliers can significantly distort the mean since it is influenced by extreme values. A single outlier with a large value can greatly pull the mean in its direction, making it an unreliable representation of the typical value in the dataset.
2. Median: The median is more robust to outliers compared to the mean. Outliers have less influence on the median since it is determined by the middle value(s) and is less affected by extreme values. Thus, the median can provide a more representative measure of central tendency in the presence of outliers.
3. Mode: Outliers generally have no effect on the mode, as it represents the most frequently occurring value(s) in the dataset. Outliers are typically less likely to be the most frequent values.

Measures of Dispersion (e.g., Range, Variance, Standard Deviation):
1. Range: Outliers can significantly affect the range since it is calculated as the difference between the maximum and minimum values. If outliers are present, they can inflate the range, making it a less reliable measure of the spread of the non-outlying values.
2. Variance and Standard Deviation: Outliers can have a substantial impact on both the variance and standard deviation since these measures are influenced by the differences between each data point and the mean. Outliers, being far away from the mean, can contribute disproportionately to the squared differences, leading to increased variance and standard deviation.
   
Example:
Consider a dataset of exam scores: [85, 92, 78, 90, 95, 100, 82, 88, 79, 250]. In this dataset, the value 250 is an outlier, significantly deviating from the other scores. Let's see how outliers affect the measures of central tendency and dispersion:

1. Mean: The presence of the outlier 250 significantly increases the mean, making it unrepresentative of the typical scores.
2. Median: The median remains unaffected by the outlier, as it is determined by the middle value(s). So, it provides a more robust measure of central tendency.
3. Mode: The mode is not affected by the outlier, as it represents the most frequent values, and the outlier does not contribute to the frequency count.
4. Range: The outlier increases the range significantly, widening the spread of the non-outlying scores.
5. Variance and Standard Deviation: The outlier greatly influences both the variance and standard deviation since they measure the differences between each score and the mean. The squared difference between the outlier and the mean contributes substantially to the increased variance and standard deviation.

Overall, outliers can distort measures of central tendency and dispersion, emphasizing the importance of considering the presence and impact of outliers when interpreting these statistical measures.

```python
d1  =  [85, 92, 78, 90, 95, 100, 82, 88, 79, 250]
d2 = [85, 92, 78, 90, 95, 100, 82, 88, 79] 

print( "d1_mean " , np.mean( d1 ) , "d2_mean " , np.mean( d2 ) )
print( "d1_mode " , stats.mode( d1 ) , "d2_mode " , stats.mode( d2 ) )
print( "d1_median " , np.median( d1 ) , "d2_median " , np.median( d2 ) )
print( "d1_std " , np.std( d1 ) , "d2_std " , np.std( d2 ) )
print( "d1_var " , np.var( d1 ) , "d2_var " , np.var( d2 ) )
```
